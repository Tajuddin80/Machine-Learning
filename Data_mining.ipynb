```python
# Missing value handling
data = [10, None, 20, 30, None]

cleaned = [x for x in data if x is not None]
print(cleaned)

mean_val = sum(cleaned) / len(cleaned)
filled = [x if x is not None else mean_val for x in data]

print("Removed:", cleaned)
print("Filled:", filled)
```

```python
# noise reduction binning method
data = [2, 4, 6, 8, 10, 12, 14]
bins = 3
bin_size = len(data) // bins
binned = []

for i in range(0, len(data), bin_size):
    group = data[i:i+bin_size]
    avg = sum(group) / len(group)
    binned.extend([avg]*len(group))

print("Smoothed data:", binned)
```

```python
# Normalization min-max

data = [10, 20, 30, 40]
min_val, max_val = min(data), max(data)
norm = [(x - min_val) / (max_val - min_val) for x in data]
print("Min-Max:", norm)
```

```python
#Normalization z score

mean = sum(data) / len(data)
std = (sum((x - mean)**2 for x in data) / len(data))**0.5
z_score = [(x - mean) / std for x in data]
print("Z-Score:", z_score)
```

```python
# pearson correlation
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

mean_x = sum(x)/len(x)
mean_y = sum(y)/len(y)

num = sum((x[i]-mean_x)*(y[i]-mean_y) for i in range(len(x)))
den = (sum((x[i]-mean_x)**2 for i in range(len(x))) * sum((y[i]-mean_y)**2 for i in range(len(y))))**0.5

r = num / den
print("Pearson Correlation:", r)
```

```python
# Cosine similarity
a = [1, 2, 3]
b = [2, 4, 6]

dot = sum(a[i]*b[i] for i in range(len(a)))
norm_a = sum(x**2 for x in a)**0.5
norm_b = sum(x**2 for x in b)**0.5
cos_sim = dot / (norm_a * norm_b)
print("Cosine Similarity:", cos_sim)
```

```python
# Visualization [text box plot style]
data = sorted([10, 12, 15, 20, 22, 23, 40])
n = len(data)
median = data[n//2]
q1 = data[n//4]
q3 = data[3*n//4]
print(f"Min={data[0]}, Q1={q1}, Median={median}, Q3={q3}, Max={data[-1]}")
```

```python
# Frequent pattern analysis
transactions = [
    {'milk', 'bread', 'butter'},
    {'milk', 'bread'},
    {'bread', 'butter'}
]

items = set(i for t in transactions for i in t)
min_support = 2
freq_items = {i: sum(1 for t in transactions if i in t) for i in items if sum(1 for t in transactions if i in t) >= min_support}

print("Frequent items:", freq_items)
```

```python
# Decision tree
def decision_tree(weather, temperature):
    if weather == "sunny" and temperature > 25:
        return "Play"
    elif weather == "rainy":
        return "Stay Home"
    else:
        return "Maybe"

print(decision_tree("sunny", 28))
print(decision_tree("rainy", 20))
```

```python
# KNN (Manual example)
data = [
    ([1, 2], 'A'),
    ([2, 3], 'A'),
    ([6, 8], 'B')
]

test = [2, 2]

def distance(p1, p2):
    return sum((p1[i]-p2[i])**2 for i in range(len(p1)))**0.5

k = 2
distances = sorted([(distance(test, x), label) for x, label in data])
neighbors = [label for _, label in distances[:k]]

# majority vote
predicted = max(set(neighbors), key=neighbors.count)
print("Predicted Class:", predicted)
```
